```python
X = df[['Annual Income (k$)','Spending Score (1-100)']]
```

```python
X.head()
```

```python
#Modelling
k=KMeans()
k.fit(X)
```

```python
k=KMeans(n_clusters=5)
k.fit_predict(X)
```

```python
#Within cluster sum of Squares
wcss=[]
```

```python
#Finding Value of K using Elbow Method
for i in range(1,11):
    k=KMeans(n_clusters=i)
    k.fit(X)
    wcss.append(k.inertia_)
```

```python
wcss
```

```python
plt.figure(figsize=(6,4))
plt.plot(range(1,11),wcss)
plt.title('Elbow Method')
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS Values')
plt.show()
```

```python
#Model Training
X = df[['Annual Income (k$)','Spending Score (1-100)']]
```

```python
k = KMeans(n_clusters=5,random_state=42)
y_means = k.fit_predict(X)
```

```python
y_means
```

```python
#Plotting clusters and cluster centroids
plt.scatter(X.iloc[y_means==0,0], X.iloc[y_means==0,1],s=80,c='red',label='Customer 1')
plt.scatter(X.iloc[y_means==1,0], X.iloc[y_means==1,1],s=80,c='yellow',label='Customer 2')
plt.scatter(X.iloc[y_means==2,0], X.iloc[y_means==2,1],s=80,c='green',label='Customer 3')
plt.scatter(X.iloc[y_means==3,0], X.iloc[y_means==3,1],s=80,c='blue',label='Customer 4')
plt.scatter(X.iloc[y_means==4,0], X.iloc[y_means==4,1],s=80,c='black',label='Customer 5')
plt.scatter(k.cluster_centers_[:,0],k.cluster_centers_[:,1],s=100,c='magenta',label='Centroids')
plt.title('Cluster of Centroids')
plt.xlabel('Annual Income (k$)')
plt.ylabel("Spending Score (1-100)")
plt.legend()
plt.show()
```
